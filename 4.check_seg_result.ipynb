```python
def searchgland(glandid):
    gland =  "./result/gland"+glandid+"_patch.png"
    glandinmsk = "./result/gland/"+glandid+"_patchinmsk.png";
    
    gland = cv2.imread(gland)
    
    glandinmsk= cv2.imread(glandinmsk)
    
    
    return gland , glandinmsk;
    
```

```python
def searchslide(slideid):
    path = "/home/ubuntu/project/pythologyData/"+slideid+"/crop.png"
    
    img = cv2.imread(path)
    
    return img
    
```

```python
slide="DSS-2018-32-002_a2"
slide = searchslide(slide)

plt.figure(figsize=(8, 8))
plt.imshow(slide)
```
